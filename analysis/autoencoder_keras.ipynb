```python
from keras.layers import Input, Dense
from keras.models import Model
import pandas as pd
from utils_analysis import normalize_dataframe2
from math import ceil
import numpy as np
from keras.models import load_model
from keras import regularizers
```

```python
data = pd.read_csv("caracteristicas.csv")
data = data.drop(data.columns[0], axis = 1)

print(data.dtypes)
```

```python
data = data.drop(labels=['id'], axis=1)
# data = data[['price', 'number_of_reviews']]
print(len(data.columns))
print(data.dtypes)
```

```python

# data, scaler = normalize_dataframe2(data)
dataset = data.to_numpy()
test_size = int(0.4 * len(dataset))
test_data = dataset[:test_size]
train_data =dataset[test_size:]
```

```python
input_dim = train_data.shape[1]
n_hidden_1 = 256
n_hidden_2 = 15
n_hidden_3 = 256

learning_rate = 0.01
n_epochs = 500
batch_size = 100
```

```python
input_l = Input(shape=(input_dim,))
encoded = Dense(n_hidden_1, activation='relu')(input_l)
encoded = Dense(n_hidden_2, activation='relu', activity_regularizer=regularizers.l1())(encoded)
# encoded = Dense(32, activation='relu')(encoded)

# decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(n_hidden_3, activation='relu')(encoded)
decoded = Dense(input_dim, activation='relu')(decoded)
```

```python
autoencoder = Model(input_l, decoded)
autoencoder.compile(optimizer='adadelta', loss='mse')
autoencoder.fit(train_data, train_data,
                epochs=n_epochs,
                batch_size=batch_size,
                shuffle=True,
                validation_data=(test_data, test_data)
                )
```

```python
prediccion = autoencoder.predict(test_data)
ind = 56
# print(prediccion[ind])
# print(test_data[ind])
i_r = prediccion
o_r = test_data
# scaler
# i_r = scaler.inverse_transform(prediccion)
# o_r = scaler.inverse_transform(test_data)
```
